### Лемматизация

```python
from pymystem3 import Mystem
m = Mystem()
purpose_unique = data['purpose'].unique()
lemmas = []
for row in purpose_unique:
    lemmas += m.lemmatize(row)
from collections import Counter
print(Counter(lemmas))
category_purpose = ['свадьба', 'недвижимость', 'автомобиль', 'образование']
def grouping_purpose(purpose):
    lemmas = m.lemmatize(purpose)
    for lemma in lemmas:
        if lemma == 'жилье':
            return 'недвижимость'
        else:
            if lemma in category_purpose:
                return lemma
data['category_purpose'] = data['purpose'].apply(grouping_purpose)
data
```

### Вывод

1. Выполнен процесс лемматизации:
 - сохранены уникальные цели получения кредита
 - проведена лемматизация для данных целей
 - леммы, на которые можно разбить цели: свадьба, недвижимость, автомобиль, образование.
2. Добавлен дополнительный столбец с категорией цели.

### Категоризация данных

```python
def grouping_children(amount):
    if amount == 0:
        return 'нет детей'
    else:
        return 'есть дети'
data['grouping_children'] = data['children'].apply(grouping_children)
def grouping_family(status):
    if status == 'гражданский брак' or status == 'женат / замужем':
        return 'семейный'
    else:
        return 'одинокий'
data['grouping_family'] = data['family_status'].apply(grouping_family)
def grouping_total_income(income):
    points = []
    quantiles = data['total_income'].quantile([0.2, 0.4, 0.6, 0.8]).astype(int)
    for row in quantiles:
        points.append(row)
    if income <= points[0]:
        return 'низкий доход'
    elif points[0] < income <= points[1]:
        return 'доход ниже среднего'
    elif points[1] < income <= points[2]:
        return 'средний доход'
    elif points[2] < income <= points[3]:
        return 'доход выше среднего'
    else:
        return 'высокий доход'
data['grouping_total_income'] = data['total_income'].apply(grouping_total_income)
data_grouping = data[['debt', 'category_purpose', 'grouping_children', 'grouping_family', 'grouping_total_income']]
data_grouping
```

### Вывод

1. Цели кредита сгруппированы в предыдущем блоке на следующие типы: 'свадьба', 'недвижимость', 'автомобиль', 'образование'.
2. По наличию детей клиенты объединены в группы 'есть дети' и 'нет детей'.
3. По семейному положению клиенты разделены на 'семейных' и 'одиноких'. К семейным отнесены клиенты замужем/женат и в гражданском браке. Остальные клиенты отнесены к одиноким.
4. По ежемесячному доходу доходы клиентов поделены на низкий доход, доход ниже среднего, средний доход, доход выше среднего, высокий доход. Разделение выполнено с помощью DataFrame.quantile. 
5. Данные группировки добавлены в общую таблицу в виде дополинтельных столбцов.
6. Для удобства новые категории клиентов и информация о задолженности вынесены в отдельную таблицу data_grouping.
7. По ежемесячному доходу доходы клиентов поделены на низкий доход, доход ниже среднего, средний доход, доход выше среднего, высокий доход. Разделение выполнено с помощью DataFrame.quantile.
